```python
count = 0
for line in book_raw:
    character_list = line.split()
    for i in target_url0:
        if(i==character_list):
            count=count+1
print("Occurrences of the word", character_list, ":", count)
```

```python
fdist = FreqDist(character_list)
```

```python
import matplotlib.pyplot as plt
fdist.plot(49,cumulative=False)
plt.show()
```
